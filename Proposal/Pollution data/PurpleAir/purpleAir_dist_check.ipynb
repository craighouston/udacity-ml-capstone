```python
import pandas as pd
# from pandas import read_csv
from pandas import datetime
import matplotlib.pyplot as plt
from matplotlib import pyplot
```

```python
purpleA = pd.DataFrame(read_csv('PurpleA/Wat Ket (18.790668 99.006142) Primary 11_25_2017 07_23_2018.csv'))
```

```python
purpleA_25 = purpleA['PM2.5_CF_1_ug/m3']
```

```python
purpleA_25.describe()
```

```python
outliers = purpleA_25 > 200
purpleA[outliers]
```

```python
ax1 = purpleA_25[purpleA_25 < 250].hist(bins=124, edgecolor='white', linewidth=.5, figsize=(20,12), grid=False)
ax1.set_xlabel('Particulate concentration (ug/m3)')
ax1.set_ylabel('Frequency')
ax1.set_facecolor("#f7f7f7")
plt.savefig('purpleA_hist.png')
```

```python
purpleA_good = purpleA_25[purpleA_25 <= 12.0]
purpleA_moderate = purpleA_25[(purpleA_25 > 12.0) & (purpleA_25 <= 35.4)]
purpleA_USG = purpleA_25[(purpleA_25 > 35.4) & (purpleA_25 <= 55.4)]
purpleA_unhealthy = purpleA_25[(purpleA_25 > 55.4) & (purpleA_25 <= 150.4)]
purpleA_vunhealthy = purpleA_25[(purpleA_25 > 150.4) & (purpleA_25 <= 250.4)]
```

```python
print(str("Good: " + "{:.2f}".format(100*(len(purpleA_good)/len(purpleA)))) + " %")
print(str("Moderate: " + "{:.2f}".format(100*(len(purpleA_moderate)/len(purpleA)))) + " %")
print(str("USG: " + "{:.2f}".format(100*(len(purpleA_USG)/len(purpleA)))) + " %")
print(str("Unhealthy: " + "{:.2f}".format(100*(len(purpleA_unhealthy)/len(purpleA)))) + " %")
print(str("Very Unhealthy: " + "{:.2f}".format(100*(len(purpleA_vunhealthy)/len(purpleA)))) + " %")
```

```python
ax3= purpleA_good.hist(bins=6, edgecolor='#f1f1f1', linewidth=.8, figsize=(20,12), grid=False, color='#00e400')
ax3= purpleA_moderate.hist(bins=11, edgecolor='#f1f1f1', linewidth=.8, figsize=(20,12), grid=False, color='#ffff00')
ax3= purpleA_USG.hist(bins=10, edgecolor='#f1f1f1', linewidth=.8, figsize=(20,8), grid=False, color='#ff7e00')
ax3= purpleA_unhealthy.hist(bins=47, edgecolor='#f1f1f1', linewidth=.8, figsize=(20,12), grid=False, color='#ff0000')
ax3= purpleA_vunhealthy.hist(bins=50, edgecolor='#f1f1f1', linewidth=.8, figsize=(20,12), grid=False, color='#99004c')

ax3.set_xlabel('Particulate concentration (ug/m3)')
ax3.set_ylabel('Frequency')
ax3.set_facecolor("#f7f7f7")
plt.savefig('purpleA_hist_color.png')
```
